```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Lionel Messi Goals.csv", encoding = "Latin1")
data
```

```python
data.isnull().sum()
```

```python
data['Date'] = data['Date'].str.replace('ï¿½', '', regex=False)
data['Date'] = data['Date'].str.extract(r'(\d{4})')

data
```

```python
goals_per_year = data['Date'].value_counts().reset_index()
goals_per_year.columns = ['Year', 'Goals']
goals_per_year = goals_per_year.sort_values(by='Year').reset_index()
goals_per_year = goals_per_year.drop(columns=['index'])

goals_per_year
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
fig.patch.set_facecolor('#2b2b2b')  
ax.set_facecolor('#2b2b2b')         

ax.bar(goals_per_year['Year'], goals_per_year['Goals'], color='#ffb6c1', alpha=0.8)


ax.set_title("Goals Scored by Messi Per Year", fontsize=16, color='#e0e0e0')
ax.set_xlabel("Year", fontsize=12, color='#e0e0e0')
ax.set_ylabel("Number of Goals", fontsize=12, color='#e0e0e0')
ax.tick_params(axis='x', labelsize=10, colors='#e0e0e0')  
ax.tick_params(axis='y', labelsize=10, colors='#e0e0e0')  

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#e0e0e0')
ax.spines['bottom'].set_color('#e0e0e0')

plt.tight_layout()
plt.show()
```

```python
penalty_goals = data[data['Goal Type'] == 'Penalty']

foot_counts = penalty_goals['Goal Method'].value_counts().reset_index()
foot_counts.columns = ['Foot', 'Count']

foot_counts
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
fig.patch.set_facecolor('#2b2b2b')  
ax.set_facecolor('#2b2b2b')         


colors = ['#87ceeb','#ffb6c1']  
wedges, texts, autotexts =ax.pie(
    foot_counts['Count'],
    labels=foot_counts['Foot'],
    autopct='%1.1f%%',
    colors=colors,
    startangle=140
)

for autotext in autotexts:
    autotext.set_color('#2b2b2b')  
    autotext.set_fontsize(10)  


for text in texts:
    text.set_color('#e0e0e0')  
    text.set_fontsize(10)
    

ax.set_title("Messi's Penalty Goals by Left and Right Foot", fontsize=12, color='#e0e0e0')

plt.tight_layout()
plt.show()
```

```python
data['Minute'] = pd.to_numeric(data['Minute'], errors='coerce')

extra_time_goals = data[data['Minute'] > 90].reset_index()
extra_time_goals = extra_time_goals.drop(columns=['index'])
extra_time_goals
```

```python
won_games = extra_time_goals[extra_time_goals['Final Game Result'] == 'Won'].reset_index().shape[0]
won_games
```

```python
total_extra_time_games = extra_time_goals.shape[0]
print(total_extra_time_games)

won_percentage = (won_games / total_extra_time_games) * 100
lost_percentage = 100 - won_percentage

results = ['Won', 'Not Won']
percentages = [won_percentage, lost_percentage]

percentages
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
fig.patch.set_facecolor('#2b2b2b')  
ax.set_facecolor('#2b2b2b')        
colors = ['#99ff99', '#ff9999']  
wedges, texts, autotexts=ax.pie(
    percentages, 
    labels=results, 
    autopct='%1.1f%%', 
    colors=colors, 
    startangle=90
)

for autotext in autotexts:
    autotext.set_color('#2b2b2b')  
    autotext.set_fontsize(10)  


for text in texts:
    text.set_color('#e0e0e0')  
    text.set_fontsize(10)

ax.set_title("Messi's extra-time goal win rate", fontsize=14, color='#e0e0e0')

plt.tight_layout()
plt.show()
```

```python
data['Messi Team'] = data.apply(
    lambda row: 'Home' if row['Home team'] in ['FC Barcelona', 'Argentina'] else 'Away', axis=1
)

data['Equalizer'] = data.apply(
    lambda row: 1 if (
        (row['Messi Team'] == 'Home' and row['Score Home Team - When Messi Scored'] == row['Score Away Team - When Messi Scored']) or
        (row['Messi Team'] == 'Away' and row['Score Away Team - When Messi Scored'] == row['Score Home Team - When Messi Scored'])
    ) else 0, axis=1
)

total_goals = data.shape[0]
equalizer_goals = data['Equalizer'].sum()


equalizer_percentage = (equalizer_goals / total_goals) * 100
non_equalizer_percentage = 100 - equalizer_percentage

print(f"Total Goals: {total_goals}")
print(f"Equalizer Goals: {equalizer_goals}")
print(f"Percentage of Equalizer Goals: {equalizer_percentage:.2f}%")
```

```python
labels = ['Equalizer Goals', 'Other Goals']
sizes = [equalizer_percentage, non_equalizer_percentage]
colors = ['#87ceeb', '#ffcccb']  


fig, ax = plt.subplots(figsize=(8, 5))
fig.patch.set_facecolor('#2b2b2b') 
ax.set_facecolor('#2b2b2b')        

ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    colors=colors,
    textprops={'color': '#e0e0e0', 'fontsize': 12},
    startangle=140
)


ax.set_title("Messi's Equalizer Goals as Percentage of Total Goals", fontsize=14, color='#e0e0e0')


plt.tight_layout()
plt.show()
```
